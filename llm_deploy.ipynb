```python
from transformers import AutoTokenizer
import pickle

# Load the tokenizer for a specific pre-trained model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Save the tokenizer to a .pkl file
with open('tokenizer.pkl', 'wb') as f:
    pickle.dump(tokenizer, f)

print("Tokenizer has been saved to tokenizer.pkl")
```

```python
import pickle

# Load the tokenizer from the .pkl file
with open('tokenizer.pkl', 'rb') as f:
    tokenizer = pickle.load(f)

print("Tokenizer has been loaded from tokenizer.pkl")
```
